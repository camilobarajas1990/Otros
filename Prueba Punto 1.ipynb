Se importan las librerías y se procede a hacer la conexión al servidor

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np

url = 'https://www.falabella.com/falabella-cl/category/cat4290063/SmartWatch?page=2'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
```

Se buscan todas las coincidencias con el marcador “a” y se mira el tamaño del data set 

```python
all_links = soup.find_all("a")
len(all_links)
```

Se crea un ciclo para traer todos los datos de la sub leyenda “ href” y se procede a visualizar la data 

```python
my_list = []

for link in all_links:
    my_list.append(link.get("href"))
print(my_list)
```

Se hace un sub set de los datos que requerimos y se procede a hacer un dataframe  

```python
my_list = my_list[913:975]
```

```python
import pandas as pd

df = pd.DataFrame({'col':my_list})
df.head()
```

Se expande el dataframe y se limpia para obtener la información de producto   

```python
new = df["col"].str.split("/", n = 30 ,expand = True) 
```

```python
Descr = new.drop_duplicates()
```

```python
Descr.shape
```

```python
Descr['Product'] = Descr[4]
```

```python
Y1 = Descr[['Product']]
Y1
```

Se procede a tomar las categorías de los head de la pagina con los Xpad empleando la librería selenium

```python
from selenium import webdriver
import pandas as pd
```

```python
driver = webdriver.Chrome()
driver.get('https://www.falabella.com/falabella-cl/category/cat4290063/SmartWatch?page=2')
```

```python
Cat_element = driver.find_elements_by_xpath('//*[@id="header"]/section[2]/div/b/span/a[1]/span')[0]
Cat = Cat_element.text
```

```python
Lin_element = driver.find_elements_by_xpath('//*[@id="header"]/section[2]/div/b/span/a[2]/span')[0]
Lin = Lin_element.text
```
